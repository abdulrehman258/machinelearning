axis='index': to drop null rows.

Rows will be dropped with missing values.

```python
df = pd.read_csv('books-data.csv')

print("before")
print(df.shape) 
print(df.isnull().sum()) 



# we will drop values with dropna
df.dropna(axis='index', inplace=True)
print("\nAfter\n")
print(df.isnull().sum())
```

axis='columns': Drops columns.

for dropping column values we can use this

```python
df = pd.read_csv('books-data.csv')

print("before")
print(df.shape) 
print(df.isnull().sum())

print("\nAfter\n")


# This will drop columns
df.dropna(axis='columns', inplace=True)
print(df.shape)
print(df.isnull().sum())
```

how='any': Remove any value that is null.

this will remove all the values with any missing values.

```python
df = pd.read_csv('books-data.csv')

print("before")
print(df.shape)
print(df.isnull().sum()) 


# Drop columns with any missing values
df.dropna(axis='columns', how='any', inplace=True)

print("\nafter\n")
print(df.shape) 
print(df.isnull().sum())
```
